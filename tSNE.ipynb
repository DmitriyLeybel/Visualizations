# t-SNE Visualization

Uses the sklearn dataset of drawn digits to test the t-SNE algorithm 

```python
%matplotlib inline
from sklearn.decomposition import TruncatedSVD
import sklearn.datasets as ds
import sklearn.manifold as mani
import matplotlib.pyplot as plt
```

```python
nums = ds.load_digits()
data = nums['data']
tsne = mani.TSNE(n_components=2)
```

Times the T-SNE transform before an initial dimension reduction

```python
%%time
redux = tsne.fit_transform(data)
```

Different colors specify different labels(0-9)

```python
plt.scatter(redux[:,0],redux[:,1], c=nums['target'],cmap='Blues')
```

```python
components = data.shape[1]/2
SVD = TruncatedSVD(n_components=int(components))
data = nums['data']
tsne = mani.TSNE(n_components=2)
```

Runs SVD first to halve the amount of dimensions, and then uses t-SNE.
No significant speed up noticed. 

```python
%%time
newdata = SVD.fit_transform(data)
redux = tsne.fit_transform(newdata)
```

```python
plt.scatter(redux[:,0],redux[:,1], c=nums['target'],cmap='Blues')
```
